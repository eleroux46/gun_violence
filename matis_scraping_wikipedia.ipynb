```python
import pandas as pd
import urllib
import bs4
```

```python
#On récupère la page des 1000 villes les plus peuplés des US
url = "https://en.wikipedia.org/wiki/List_of_United_States_cities_by_population"
response = urllib.request.urlopen(url).read()
page = bs4.BeautifulSoup(response)
```

```python
tableau = page.find("table", {"class" : "wikitable sortable"}) #on récupère le tableau désiré
table_body = tableau.find("tbody")
rows = table_body.find_all("tr")
```

```python
dic_villes = dict()
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    if len(cols) > 0 : 
        dic_villes[cols[0]] = cols[1:]

villes_df = pd.DataFrame.from_dict(dic_villes, orient="index").reset_index()
villes_df.columns = ["city_name", "state", "2022 estimate", "2020_census", "change", "2020 land area (sq mi)", "2020 land area (km^2)", "2020 population density (/sq mi)", "2020 population density (/km^2)", "location" ]
villes_df.set_index("city_name")
```
